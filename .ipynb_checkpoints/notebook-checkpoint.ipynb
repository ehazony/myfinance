```python
import os
import plotly.graph_objects as go
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "finance.settings")
import django
django.setup()
os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
import plotly
# plotly.offline.init_notebook_mode(connected=True)
from myFinance.models import *
import plotly.express as px
import pandas as pd
import tabulate
from itertools import groupby
from django_pivot.histogram import histogram
TAG_CHOICES = [tag.name for tag in Tag.objects.all()]
```

```python
def bar_fig_by_month(transactions):
    ts_pd = pd.DataFrame(list(transactions))
    ts_pd
    fig = px.bar(ts_pd, x='month_date', y = 'value__sum',color='value__sum')
    fig.show()
    
def monthly_by_name(_name):
    transaction_sum =Transaction.objects.all().filter(name = _name).values('month_date').annotate(Sum('value'))
    bar_fig_by_month(transaction_sum)
    
def monthly_by_tag(_tag):
    transaction_sum =Transaction.objects.all().filter(tag = _tag).values('month_date').annotate(Sum('value'))
    bar_fig_by_month(transaction_sum)

def monthly_average_by_category():
    trans = Transaction.objects.all().filter(month__in=[12,1,2,3]).values('tag_ref__name').annotate(Sum('value'))
    trans= pd.DataFrame(list(trans))
    trans["value__sum"] = trans["value__sum"]/4
    trans = trans.rename(columns={ 'tag_ref__name' : 'tag', 'value__sum': 'monthly average'}, index={'ONE': 'one'})
    trans = trans [[ 'monthly average', 'tag']]
    return trans
    print(tabulate(trans, headers='keys', tablefmt='psql'))

def transactions_df():
    transactions = Transaction.objects.all().order_by('date')
    transactions_df = pd.DataFrame(list(transactions.values()))
    return transactions_df

def transactions_set():
    transactions = Transaction.objects.all().order_by('date')
    return transactions

def to_df(transactions_set):
    return pd.DataFrame(list(transactions_set.values()))
```

```python
monthly_average_by_category()
```

```python
transactions_set = transactions_set()
transactions_set= transactions_set.all().filter(tag = 'same')
transactions_df = to_df(transactions_set)
fig = px.scatter(transactions_df,x= "date", y="value",
                 hover_name="tag", hover_data=["name"])
fig.show()
```

```python
# only works for one month. 
trans = Transaction.objects.all().filter(month__in=[12]).order_by('date').values('name','month','date','value')
df= pd.DataFrame(list(trans))
df['date'] = df['date'].astype('datetime64[ns]')
gb= df.set_index('date').groupby([pd.Grouper(freq='D'),'month']).sum().cumsum()
fig = px.line(gb,x=df.date.unique(),  y="value")


trans = Transaction.objects.all().filter(month__in=[1]).order_by('date').values('month','date','value')
df= pd.DataFrame(list(trans))
gb= df.groupby(["month","date"]).sum().cumsum()
fig1 =  px.line(gb,x=df.date.unique(),  y="value")
fig.add_trace(fig1.data[0])

trans = Transaction.objects.all().filter(month__in=[2]).order_by('date').values('month','date','value')
df= pd.DataFrame(list(trans))
gb= df.groupby(["month","date"]).sum().cumsum()
fig1 =  px.line(gb,x=df.date.unique(),  y="value")
fig.add_trace(fig1.data[0])

trans = Transaction.objects.all().filter(month__in=[3]).order_by('date').values('month','date','value')
df= pd.DataFrame(list(trans))
gb= df.groupby(["month","date"]).sum().cumsum()
fig1 =  px.line(gb,x=df.date.unique(),  y="value")
fig.add_trace(fig1.data[0])
fig.show()

```
